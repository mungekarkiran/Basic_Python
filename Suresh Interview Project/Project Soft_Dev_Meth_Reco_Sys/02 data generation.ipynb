```python
import pandas as pd
import numpy as np

import seaborn as sns
```

```python
data = pd.read_excel('dataset\Software developement methodology recommender system updated_20231123.xlsx', sheet_name='Sheet2')
data
```

```python
# data.columns
# https://stackoverflow.com/questions/13411544/delete-a-column-from-a-pandas-dataframe
data.drop('Requirements are Partially defined early in the SDLC.1', axis=1, inplace=True)
data.columns
```

```python
# https://stackoverflow.com/questions/11346283/renaming-column-names-in-pandas
# https://www.programiz.com/python-programming/examples/list-to-dictionary
col_list = list(data.columns)
new_col_list = [f'Q_{i+1}' for i in range(len(col_list))]
col_dict = dict(map(lambda i,j : (i,j) , col_list,new_col_list))

data.rename(columns=col_dict, inplace=True)
```

```python
data['traditional_approach'] = np.where(
    (data['Q_1'] == 0) |
    (data['Q_2'] == 1) |
    (data['Q_3'] == 1) |
    (data['Q_4'] == 0) |
    (data['Q_5'] == 1) |
    (data['Q_6'] == 1) |
    (data['Q_7'] == 0) |
    (data['Q_8'] == 1) |
    (data['Q_9'] == 1) |
    (data['Q_10'] == 1) |
    (data['Q_11'] == 1) |
    (data['Q_12'] == 0) |
    (data['Q_13'] == 0) |
    (data['Q_14'] == 0) |
    (data['Q_15'] == 0) |
    (data['Q_16'] == 1) |
    (data['Q_17'] == 1) |
    (data['Q_18'] == 1) |
    (data['Q_19'] == 0) |
    (data['Q_20'] == 0) |
    (data['Q_21'] == 1) |
    (data['Q_22'] == 0) |
    (data['Q_23'] == 0) |
    (data['Q_24'] == 0), '1', '0')
data
```

```python
data['agile_approach'] = np.where(
    (data['Q_1'] == 0) |
    (data['Q_2'] == 1) |
    (data['Q_3'] == 0) |
    (data['Q_4'] == 1) |
    (data['Q_5'] == 1) |
    (data['Q_6'] == 0) |
    (data['Q_7'] == 1) |
    (data['Q_8'] == 1) |
    (data['Q_9'] == 1) |
    (data['Q_10'] == 0) |
    (data['Q_11'] == 0) |
    (data['Q_12'] == 1) |
    (data['Q_13'] == 1) |
    (data['Q_14'] == 1) |
    (data['Q_15'] == 0) |
    (data['Q_16'] == 1) |
    (data['Q_17'] == 0) |
    (data['Q_18'] == 1) |
    (data['Q_19'] == 1) |
    (data['Q_20'] == 1) |
    (data['Q_21'] == 0) |
    (data['Q_22'] == 1) |
    (data['Q_23'] == 1) |
    (data['Q_24'] == 1), '1', '0')
data
```

```python
data['approach_type'] = data['traditional_approach'] + data['agile_approach'] 
data
```

```python
data['approach'] = np.where(
    (data['Requirements are defined early in the SDLC']=='1' )
    | (data['Requirements are easily defined and understandable']=='1')
    | (data['Change Management_in the early process']=='1')
    | (data['Limited User Involvement']=='1')
    | (data['No Experience of Participating in Similar Projects']=='1')
    | (data['Little Experience of Similar Projects']=='1')
    | (data['Little Domain Knowledge_new to Technology']=='1')
    | (data['Little experience on Tools']=='1')
    | (data['Improvement of an Old System']=='1')
    | (data['Stable Funding']=='1')
    | (data['Highly Reliable Requirements']=='1')
    | (data['Tight Schedule']=='1')
    | (data['Scarce Resources']=='1'), '1', '0')
data
```

```python
data['Prototyping Model'] = np.where(
    (data['Requirements are changed frequently']=='1' )
    | (data['Change Management_in the early process']=='1')
    | (data['User Participation in All Phases']=='1')
    | (data['Little Experience of Similar Projects']=='1')
    | (data['Little Domain Knowledge_new to Technology']=='1')
    | (data['Little experience on Tools']=='1')
    | (data['Training Availability when needed']=='1')
    | (data['Reusable Components']=='1'), '1', '0')
data
```

```python
data['Rapid Application Development'] = np.where(
    (data['Requirements are changed frequently']=='1' )
    | (data['Change Management_in the early process']=='1')
    | (data['User Participation in All Phases']=='1')
    | (data['No Experience of Participating in Similar Projects']=='1')
    | (data['Little Domain Knowledge_new to Technology']=='1')
    | (data['Training Availability when needed']=='1')
    | (data['Tight Schedule']=='1')
    | (data['Reusable Components']=='1'), '1', '0')
data
```
