# generate dataset for 15Q?

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
col_name = [f'Q_{i+1}' for i in range(15)]
col_len = len(col_name)
data = pd.DataFrame(columns=col_name)

for i in tqdm(range(2**col_len)):
    data.loc[len(data.index)] = [*(bin(i).split('b')[1].zfill(col_len))]
data
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data
```

```python
data['waterfall_approach'] = np.nan
data['agile_approach'] = np.nan
data['final_approach'] = np.nan
data
```
