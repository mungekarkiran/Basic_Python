```python
image_dir = os.path.join('..', 'Dataset', 'CUB_Dataset', 'CUB_200_2011', 'images')

for img_folder in glob.glob(os.path.join(image_dir, '*')):
    print(img_folder.split('\\')[-1])
    class_name = img_folder.split('\\')[-1]
    
    if class_name in list_of_class_0:
        # Define the source folder path and the destination directory path
        source_folder = glob.glob(os.path.join(img_folder, '*'))
        destination_directory = os.path.join('..', 'Dataset', 'CUB_Dataset', 'Primary_classes', '0')
        
        if not os.path.exists(destination_directory):
            os.makedirs(destination_directory)

        # Copy the entire files to the destination directory
        try:
            for source_file in source_folder:
                shutil.copy(source_file, destination_directory)
#                 sys.stdout.write(f"Folder copied successfully from {source_file} to {destination_directory}")
#                 sys.stdout.flush()
        except Exception as e:
            pass

    elif class_name in list_of_class_1:
        # Define the source folder path and the destination directory path
        source_folder = glob.glob(os.path.join(img_folder, '*'))
        destination_directory = os.path.join('..', 'Dataset', 'CUB_Dataset', 'Primary_classes', '1')

        if not os.path.exists(destination_directory):
            os.makedirs(destination_directory)

        # Copy the entire folder to the destination directory
        try:
            for source_file in source_folder:
                shutil.copy(source_file, destination_directory)
#                 sys.stdout.write(f"Folder copied successfully from {source_file} to {destination_directory}")
#                 sys.stdout.flush()
        except Exception as e:
            pass

    else:
        pass
#     break
```
