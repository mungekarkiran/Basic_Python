# To rename the folders 

````python
import os

def rename_folders_in_directory(parent_directory):
    for folder_name in os.listdir(parent_directory):
        folder_path = os.path.join(parent_directory, folder_name)

        # Check if it's a directory
        if os.path.isdir(folder_path):
            # Split the folder name at the first dot and take the part after it
            new_name = folder_name.split('.', 1)[-1] if '.' in folder_name else folder_name
            new_folder_path = os.path.join(parent_directory, new_name)

            # Rename the folder
            if new_folder_path != folder_path:
                os.rename(folder_path, new_folder_path)
                print(f"Renamed: {folder_path} -> {new_folder_path}")
            else:
                print(f"No renaming needed for: {folder_path}")


if __name__ == "__main__":
    # parent_directory = os.path.join('..','Dataset', 'CUB_dataset_12_june_2024', 'archive', 'CUB_reduced_dataset_main')
    parent_directory = os.path.join('..','Dataset', 'CUB_dataset_12_june_2024', 'archive', 'Test_class_glove')
    rename_folders_in_directory(parent_directory)
````

# To combine csv #

````python
import pandas as pd
import glob

def combine_csv_rowwise(input_folder, output_file):
    # Get a list of all CSV files in the folder
    csv_files = glob.glob(f"{input_folder}/*.csv")
    print(f"Found {len(csv_files)} CSV files.")

    # Initialize an empty DataFrame
    combined_df = pd.DataFrame()

    for file in csv_files:
        # Read each CSV file into a DataFrame
        df = pd.read_csv(file)
        combined_df = pd.concat([combined_df, df], ignore_index=True)  # Append rows to the combined DataFrame

    # Save the combined DataFrame to a new CSV file
    combined_df.to_csv(output_file, index=False)
    print(f"Combined CSV saved to {output_file}")

# Usage Example
input_folder = r"D:\ZSL\CUB_dataset_codes\Hybrid_feature_approach_CUB_BERT\Hybrid_approach_CUB_BERT"  # Replace with the folder containing your CSV files
output_file = "Hyrid_CUB_BERT_result.csv"  # Replace with the desired output file name
combine_csv_rowwise(input_folder, output_file)
````

### Option 2: Train Your Own Word2Vec Model

If you prefer to train your own Word2Vec embeddings for a custom dataset:

#### Steps to Train Word2Vec

1. Prepare a Text Corpus:
    - Collect a large corpus of text data related to your domain. Save it as a .txt file.

2. Install Gensim:
    
    `
    pip install gensim
    `

3. Train the Word2Vec Model: Use the gensim library to train a Word2Vec model.

    ```Python
    from gensim.models import Word2Vec

    # Load your text corpus
    corpus_file = 'your_text_corpus.txt'  # Replace with your text corpus file
    with open(corpus_file, 'r', encoding='utf-8') as f:
        sentences = [line.strip().split() for line in f]  # Tokenize sentences

    # Train Word2Vec model
    model = Word2Vec(
        sentences,
        vector_size=300,  # Number of dimensions for the embeddings
        window=5,         # Context window size
        min_count=5,      # Minimum word frequency to include in the vocabulary
        workers=4         # Number of threads
    )

    # Save the model in binary format
    model.wv.save_word2vec_format('custom_word2vec.bin', binary=True)
    print("Word2Vec model saved as 'custom_word2vec.bin'")
    ```

````python
import os
from gensim.models import Word2Vec

corpus_file_path = os.path.join('..', 'Dataset', 'CUB_dataset_12_june_2024', 'archive', 'corpus', 'CUB_corpus_Test.txt')
save_word2vec_file_path = os.path.join('..', 'Dataset', 'CUB_dataset_12_june_2024', 'archive', 'corpus', 'custom_word2vec_Test.bin')

# Load your text corpus
corpus_file = corpus_file_path  # Replace with your text corpus file
with open(corpus_file, 'r', encoding='utf-8') as f:
    sentences = [line.strip().split() for line in f]  # Tokenize sentences
    
# print(f"sentences : {sentences}")

# Train Word2Vec model
model = Word2Vec(
    sentences,
    vector_size=300,  # Number of dimensions for the embeddings
    window=2,         # Context window size
    min_count=1,      # Minimum word frequency to include in the vocabulary ()
    workers=4         # Number of threads
)

# Build the vocabulary
model.build_vocab(sentences)

# Train the Word2Vec model
model.train(sentences, 
            total_examples=model.corpus_count, 
            epochs=1000, # model.epochs, 
            start_alpha=(model.min_alpha + (53/200 * (model.alpha - model.min_alpha)))
           )

# Save the model in binary format
model.wv.save_word2vec_format(save_word2vec_file_path, binary=True)
print(f"{model} : Word2Vec model saved as 'custom_word2vec.bin'")
````

### Get top similar words

````python
from gensim.models import KeyedVectors

def get_top_similar_words(model_path, word, top_n=10):
    """
    Load a Word2Vec model and get the top N similar words for a given input word.
    
    Parameters:
        model_path (str): Path to the pre-trained Word2Vec binary model.
        word (str): The word for which to find similar words.
        top_n (int): Number of similar words to retrieve (default is 10).
        
    Returns:
        list of tuples: List of top N similar words with their similarity scores.
    """
    try:
        # Load the Word2Vec model
        print(f"Loading model from: {model_path}")
        model = KeyedVectors.load_word2vec_format(model_path, binary=True)
        print(f"{model} : Model loaded successfully.")
        
        # Get top N similar words
        if word in model.key_to_index:
            similar_words = model.most_similar(word, topn=top_n)
            return similar_words
        else:
            print(f"The word '{word}' is not in the model's vocabulary.")
            return []
    except Exception as e:
        print(f"Error: {e}")
        return []
````

````python
# Example: Replace with the actual path to your model
model_path = save_word2vec_file_path # 'GoogleNews_vectors_negative300/corpus/custom_word2vec.bin'

# Input word to find similar words for
input_word = 'Parakeet_Auklet' # 'Ivory_Gull' 'Black_footed_Albatross''California_Gull''Crested_Auklet','Least_Auklet','Brandt_Cormorant'
                                # 'Yellow_breasted_Chat','Brewer_Blackbird','Black_billed_Cuckoo','Scarlet_Tanager','Cardinal'
                                #'Green_Kingfisher'
# Get top 10 similar words
similar_words = get_top_similar_words(model_path, input_word, top_n=10)

if similar_words:
    print(f"Top 10 words similar to '{input_word}':")
    for word, score in similar_words:
        print(f"{word}: {score}")
````
