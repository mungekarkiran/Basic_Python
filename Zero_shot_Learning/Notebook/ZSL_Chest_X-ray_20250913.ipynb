```python
import os
import numpy as np
import pandas as pd 
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import gensim
import gensim.downloader as gdownloader
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
```

```python


# Load ResNet101 model
resnet = tf.keras.applications.ResNet101(include_top=False, pooling='avg', input_shape=(224, 224, 3))
resnet.trainable = False  # Freeze the ResNet layers for feature extraction

def extract_features(image_dir, class_list):
    features = []
    labels = []
    for class_name in class_list:
        class_dir = os.path.join(image_dir, class_name)
        for img_file in os.listdir(class_dir):
            img_path = os.path.join(class_dir, img_file)
            img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
            img_array = tf.keras.preprocessing.image.img_to_array(img)
            img_array = tf.keras.applications.resnet.preprocess_input(img_array)  # Preprocess image for ResNet
            img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
            feature = resnet.predict(img_array)
            features.append(feature.squeeze())  # Remove the batch dimension (output of ResNet)
            labels.append(class_name)
    return np.array(features), np.array(labels)
```

```python
import random
from collections import defaultdict

def create_sampled_pairs(visual_features, labels, class_to_semantic, num_pos=5, num_neg=5):
    pairs = []
    pair_labels = []

    # Group image indices by class
    class_to_indices = defaultdict(list)
    for idx, label in enumerate(labels):
        class_to_indices[label].append(idx)

    all_indices = list(range(len(labels)))

    for idx, label in enumerate(labels):
        visual = visual_features[idx]
        semantic = class_to_semantic[label]

        # Positive pairs
        pos_indices = class_to_indices[label].copy()
        pos_indices.remove(idx)  # don't pair with self
        pos_indices = random.sample(pos_indices, min(num_pos, len(pos_indices)))

        for p_idx in pos_indices:
            pairs.append([visual, visual_features[p_idx], semantic, class_to_semantic[labels[p_idx]]])
            pair_labels.append(1)

        # Negative pairs
        neg_classes = [cls for cls in class_to_indices if cls != label]
        for _ in range(min(num_neg, len(neg_classes))):
            neg_class = random.choice(neg_classes)
            neg_idx = random.choice(class_to_indices[neg_class])
            pairs.append([visual, visual_features[neg_idx], semantic, class_to_semantic[labels[neg_idx]]])
            pair_labels.append(0)

#     return np.array(pairs), np.array(pair_labels)
    return pairs, pair_labels
```

```python
# def build_siamese_model(visual_dim=2048, semantic_dim=300, embed_dim=512):
#     visual_input = layers.Input(shape=(visual_dim,), name='visual_input')
#     visual_branch = layers.Dense(embed_dim, activation='relu')(visual_input)
#     visual_branch = layers.BatchNormalization()(visual_branch)

#     semantic_input = layers.Input(shape=(semantic_dim,), name='semantic_input')
#     semantic_branch = layers.Dense(embed_dim, activation='relu')(semantic_input)
#     semantic_branch = layers.BatchNormalization()(semantic_branch)

#     normalized_visual = tf.nn.l2_normalize(visual_branch, axis=1)
#     normalized_semantic = tf.nn.l2_normalize(semantic_branch, axis=1)
#     cosine_sim = tf.reduce_sum(normalized_visual * normalized_semantic, axis=1, keepdims=True)

#     model = Model(inputs=[visual_input, semantic_input], outputs=cosine_sim)
#     return model


def build_siamese_model(visual_dim=2048, semantic_dim=300, embed_dim=512):
    # Visual branch
    visual_input = layers.Input(shape=(visual_dim,), name='visual_input')
    v = layers.Dense(1024, activation='relu')(visual_input)
    v = layers.BatchNormalization()(v)
    v = layers.Dropout(0.3)(v)
    v = layers.Dense(embed_dim, activation='relu')(v)
    v = layers.BatchNormalization()(v)
    normalized_visual = tf.nn.l2_normalize(v, axis=1)

    # Semantic branch
    semantic_input = layers.Input(shape=(semantic_dim,), name='semantic_input')
    s = layers.Dense(512, activation='relu')(semantic_input)
    s = layers.BatchNormalization()(s)
    s = layers.Dropout(0.3)(s)
    s = layers.Dense(embed_dim, activation='relu')(s)
    s = layers.BatchNormalization()(s)
    normalized_semantic = tf.nn.l2_normalize(s, axis=1)

    # Cosine similarity
    cosine_sim = tf.reduce_sum(normalized_visual * normalized_semantic, axis=1, keepdims=True)

    model = Model(inputs=[visual_input, semantic_input], outputs=cosine_sim)
    return model

class MarginCosineLoss(tf.keras.losses.Loss):
    def __init__(self, margin=0.4):
        super().__init__()
        self.margin = margin

    def call(self, y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        pos_loss = y_true * tf.square(1 - y_pred)
        neg_loss = (1 - y_true) * tf.square(tf.maximum(0.0, y_pred - self.margin))
        return tf.reduce_mean(pos_loss + neg_loss)


# # Optional: Custom cosine similarity loss with margin
# class CosineSimilarityLoss(tf.keras.losses.Loss):
#     def __init__(self, margin=0.3):
#         super().__init__()
#         self.margin = margin

#     def call(self, y_true, y_pred):
#         y_true = tf.cast(y_true, tf.float32)
#         pos_loss = y_true * (1 - y_pred)
#         neg_loss = (1 - y_true) * tf.maximum(0.0, y_pred - self.margin)
#         return tf.reduce_mean(pos_loss + neg_loss)

# Build and compile model
model = build_siamese_model()
model.compile(optimizer='adam', loss=MarginCosineLoss())


# Extract input arrays from pairs and convert to float32
visual_inputs = np.stack(pairs[:, 0]).astype('float32')
semantic_inputs = np.stack(pairs[:, 2]).astype('float32')
pair_labels = pair_labels.astype('float32')  # Optional: ensure consistent type




callbacks = [
    ReduceLROnPlateau(monitor='loss', factor=0.5, patience=10, min_lr=1e-6),
    EarlyStopping(monitor='loss', patience=20, restore_best_weights=True)
]


model.fit(
    x={'visual_input': visual_inputs, 'semantic_input': semantic_inputs},
    y=pair_labels,
    batch_size=32,
    epochs=300,
    callbacks=callbacks,  # 👈 Add this line
    shuffle=True
)


# # Now train the model
# model.fit(
#     x={'visual_input': visual_inputs, 'semantic_input': semantic_inputs},
#     y=pair_labels,
#     batch_size=8,
#     epochs=10


```
